## Linear Regression

```python
import pandas as pd
df = pd.read_csv("Stock_data.csv")

x = df[['Volatility', 'Market Return']]
y = df['Stock return']

from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```

## Lasso Regression

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Stock_data.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(['Stock'], axis = 1, inplace = True)
df
```

```python
x = df[['Volatility', 'Market Return']]
y = df['Stock return']
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
x_pred_train = lr.predict(x_train) 
r2 = r2_score(y_train, x_pred_train)
```

```python
print(r2)
```

```python
x_pred_test = lr.predict(x_test)
R2 = r2_score(y_test, x_pred_test)
```

```python
print(R2)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(x_train, y_train)
```

```python
x_pred_lasso_test = lasso.predict(x_test)
r2_score(y_test, x_pred_lasso_test)
```
